```python
def check_even_list(num_list):
    
    even_numbers = []
    
    # Go through each number
    for number in num_list:
        # Once we get a "hit" on an even number, we append the even number
        if number % 2 == 0:
            even_numbers.append(number)
        # Don't do anything if its not even
        else:
            pass
    # Notice the indentation! This ensures we run through the entire for loop    
    return even_numbers
```

```python
class Number:
    def __init__(self, Number):
        self.Number = Number

    def Display(self):
        print(self.Number)


class Palindrome(Number):
    def __init__(self, Number):
        super().__init__(Number)  # Initialize the parent class

    def Is_palindrome(self):
        num_str = str(self.Number)
        return num_str == num_str[::-1]


class Prime(Number):
    def __init__(self, Number):
        super().__init__(Number)  # Initialize the parent class

    def Is_prime(self):
        if self.Number <= 1:
            return False
        for i in range(2, int(self.Number**0.5) + 1):
            if self.Number % i == 0:
                return False
        return True


class Armstrong(Number):
    def __init__(self, Number):
        super().__init__(Number)  # Initialize the parent class

    def Is_armstrong(self):
        num_str = str(self.Number)
        length = len(num_str)
        total = 0
        for digit in num_str:
            total += int(digit) ** length
        return total == self.Number


# Take input from the user
num = int(input("Enter a number: "))

# Create instances of Palindrome, Prime, and Armstrong
palindrome_checker = Palindrome(num)
prime_checker = Prime(num)
armstrong_checker = Armstrong(num)

# Check if the number is a palindrome
if palindrome_checker.Is_palindrome():
    print(f"'{num}' is a palindrome.")
else:
    print(f"'{num}' is not a palindrome.")

# Check if the number is prime
if prime_checker.Is_prime():
    print(f"'{num}' is a prime number.")
else:
    print(f"'{num}' is not a prime number.")

# Check if the number is an Armstrong number
if armstrong_checker.Is_armstrong():
    print(f"'{num}' is an Armstrong number.")
else:
    print(f"'{num}' is not an Armstrong number.")
```
