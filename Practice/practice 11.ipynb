# line plot

```python
x_axis = [1,2,3]
y_axis = [4,5,6]

#line plot
plt.plot(x_axis,y_axis)
plt.show() # if using .py file
```

```python
plt.title('LINEPLOT') # all the commands must be in single shell
plt.plot(x_axis,y_axis,marker = 'x')
plt.show()
```

```python
plt.title('LINEPLOT') # all the commands must be in single shell
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x_axis,y_axis,marker = 'x')
plt.show()
```
