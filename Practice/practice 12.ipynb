# creating multiple subplots in one figure

```python
x1_axis=[2,4,5,6]
y1_axis=[1,10,100,100]

x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]
```

```python
# help(plt.subplot)
```

```python
#plt.subplot(no of rows,no of columns,current plot no)
plt.subplot(1,2,1)
plt.plot(x1_axis,y1_axis,'r')
plt.subplot(1,2,2)
plt.plot(x2_axis,y2_axis,'g')

plt.tight_layout() # to show in sub plot
```

```python
plt.subplot(1,2,1)
plt.title('EXPONENTIAL')
plt.plot(x1_axis,y1_axis,'r')
plt.subplot(1,2,2)
plt.title('LINEAR')
plt.plot(x2_axis,y2_axis,'g')

plt.tight_layout() # to show in sub plot
```

# Load Dataset

```python
import seaborn as sns
import pandas as pd
dataset = pd.read_csv('data.csv')
```

# statistical relationship - scatter plot

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset) # Assigns different colour to different age groups
```

```python
sns.relplot(x='hours',y='marks',data=dataset) # same colour
```

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,kind="line",style="internet")
```

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,col="internet") # cfeates two different plots for 0 and 1
```

```python
sns.relplot(x='hours',y='marks',hue='internet',data=dataset,col="age") # creates 6 different plots
```

```python
# DATA DISTRIBUTION - HISTOGRAM
```

```python
sns.displot(dataset,x="age",binwidth=0.5,hue="internet")
```

```python
sns.displot(dataset,x="age",col="internet") #Multiple Data Distribution
```

```python
sns.displot(dataset,x="age",y="marks",hue="internet") #creates Bivariant plot
```

```python
sns.displot(dataset,x="age",y="marks",kind="kde",hue="internet")#density plot 
```

```python
sns.pairplot(dataset) # compares all the values
```
